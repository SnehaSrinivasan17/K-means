```python
""" 
K means clustering algorithm
K means is an unsupervised algo (finds patterns in the dataset and not predictions).
It is an iterative algorithm.

Steps:
1. Specify the number of clusters (k)
2. Randomly initialize the centroid for each cluster
3. Data points are appointed to a cluster based on it's distance from the centroid
4. Update the centroids based on the geometric mean of the data points
5. Repeat until same centroids are repeated

"""
```

```python
import pandas as pd
import numpy as np
```

```python
# load player data
players = pd.read_csv("C:/Users/sneha/Downloads/players_22.csv")
```

```python
players
```

```python
# features based on which, clustering happens
features = ["overall","potential","wage_eur","value_eur","age"]
```

```python
# drop missing or NA values from the selcted columns (clustering usually doesnt work with them)
players = players.dropna(subset = features)
```

```python
data = players[features].copy()
```

```python
data
```

```python
"""
1. Scale the data (to set a range for the values in the column, to prevent one column from dominating)
2. Initialize random centroids
3. Label each data point based on distance from centroids
4. Update centroids
5. Repeat till convergence

"""
```

```python
# 1. Scale the data (make sure no negative vals or 0 is there) [range from 1 to 10]
data = ((data - data.min()) / (data.max() - data.min())) * 10 + 1
```

```python
data
```

```python
data.describe()
```

```python
data.head()
```

```python
# 2.Intialize the random centroids randomly
# series = array
def random_centroids(data,k):
    centroids = [] #series
    for i in range (k):
        centroid = data.apply(lambda x: float(x.sample())) #goes through each value in a column and a series is returned which is converted to a float)
        centroids.append(centroid) #adding into the series
    return pd.concat(centroids,axis=1) #combines all the series to one df
```

```python
centroids = random_centroids(data, 5)
```

```python
centroids
```

```python
# 3. Distance between center and data point
def get_labels(data,centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
    return distances.idxmin(axis = 1) #index of the min value = cluster value
```

```python
labels = get_labels(data, centroids)
```

```python
labels.value_counts() #number of times each unique value occurs
```

```python
# 4. Update the cluster centroids
def new_centroids(data,labels,k):
    centroids = data.groupby(labels).apply(lambda x:np.exp(np.log(x).mean())).T #geometric mean
    return centroids
    
    #splitting the df data by labels into clusters and then finding teh geometric mean for each cluster
```

```python
# Plot the cluster assignments at each iteration
from sklearn.decomposition import PCA #summarize 5 columns to 2
import matplotlib.pyplot as plt
from IPython.display import clear_output #clear the output in the notebook on each iteration
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True) #clears the graph displayed
    plt.title(f'Iteration{iteration}') #title of the graph
    plt.scatter(x=data_2d[:,0],y=data_2d[:,1], c=labels) #colored according to labels
    plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1]) #plotting on top of data
    plt.show()
```

```python
# 5.loop to convergence
max_iterations = 100
k = 3
centroids = random_centroids(data,k) #initialise random centroids
old_centroids = pd.DataFrame() #same at the start
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids
    
    labels = get_labels(data,centroids)
    centroids = new_centroids(data,labels,k)
    plot_clusters(data,labels,centroids,iteration)
    iteration+=1
```
